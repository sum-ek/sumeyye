```r
titanic %>% group_by(pclass) %>% summarise(mean_age= mean(age, na.rm=T))
```

```r
titanic_at <- titanic %>% na.omit%>% group_by(pclass, survived)  %>% summarise(mean_age= mean(age, na.rm=T))
titanic_at
```

```r
titanic_at_wide <- titanic_at %>% pivot_wider(id_cols=pclass, names_from=survived, values_from=mean_age)
titanic_at_wide
```

```r
titanic_at_wide %>% pivot_longer(cols= -"pclass", names_to="survived2", values_to= "mean_age2", values_drop_na=T)
```
