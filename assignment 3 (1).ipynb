```r
hamlet$dialogue %>% unlist %>% str_extract_all("thee") # from the hamlet data's dialogue variable, i extracted the lines that have "thee" word.
```

```r
hamlet$dialogue %>% unlist %>% str_subset("\\?$") # i found the question sentences from the dialogue
```

```r
hamlet %>% filter(act=="Act1") %>% unlist %>% str_subset("^What") #I printed the dialogues start with "what" from the act 1
```

```r
hamlet %>% filter(act=="Act1"& scene=="Scene I"& person== "Bernardo") %>% unlist %>% str_subset("^Wh") #from act 1 scene 1 Bernardo's sentences starts with wh(question words)
```

```r
hamlet$dialogue %>% unlist %>% str_subset("\\d+") #i wondered is there any digit in the dialogues, there is not
```

```r
hamlet %>% filter(person=="Hamlet") %>% unlist %>% str_subset("\\bthou\\b") #filtered hamlet's dialogues that have the "thou" word
```

```r
new_hamlet <- hamlet$dialogue %>% str_replace("\\bthou\\b", "you") %>% str_replace("\\bthy\\b", "your")
new_hamlet #i replaced the old words with modern words
```

```r
keyboard_f <-".çömnbvcxz< ,işlkjhgfdsa üğpoıuytrewq -*0987654321"
```

```r
keyboard_f %>% str_extract_all("[^\\d]") #i excluded the digits from my data
```

```r
keyboard_f %>% str_extract_all("ç*ö[a-z]+") #starts with ç or not, continues with ö and continues with a to z (charachers)
```

```r
keyboard_f %>% str_extract_all("<?") #results have zero or one "<"
```

```r
keyboard_f %>% str_extract_all("ü.+q") #starts with ü and anything and one or more things and ends with q
```

```r
keyboard_f %>% str_replace_all("\\s+", "") #boşlukların yerine hiçbir şey koyarak boşlukları kapatmış oldum
```

```r
keyboard_f %>% str_extract_all("\\w+") #bir veya daha fazla alphanumeric karakter içerenleri çıkartmış oldum
```

```r
keyboard_f %>% str_extract_all("\\d+") #bir veya daha fazla digit içeren bi sonuç çıkartmış oldum
```

```r
keyboard_f %>% str_extract_all("([a-z]|\\d)+") #a to z içeren veya digit içeren sonuçları çıkarttım 
```
